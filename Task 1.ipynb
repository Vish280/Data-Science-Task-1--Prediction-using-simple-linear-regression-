TASK - Predicting percentage of student based on no. of hours studied using simple linear regression

```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

STEP 1 - Reading data using it's URL

```python
dataset = pd.read_csv("http://bit.ly/w-data")
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
dataset.describe()
```

STEP 2 - Visualizing the input data

```python
#ploting graph of no. of hours studied by the user to the percentage that will be scored
dataset.plot(x= "Hours", y="Scores",style= "o" ,c="g")
plt.xlabel("Hours Studied")
plt.ylabel("Score(in %)")
plt.title("Hours VS Scores")
plt.show()
```


STEP 3 - Processing the data

```python
#data is divided into 'attributes' and 'label' using Slicing
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values
```

STEP 4 - Training our model

```python
#Splitting the dataset into training and testing models.
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2, random_state= 0 )
lr = LinearRegression()
lr.fit(x_train, y_train)
print("Training has been successfully completed!")
```

STEP 5 - Plotting the line of regression

```python
line = lr.coef_* x + lr.intercept_  
plt.scatter(x,y)
plt.plot(x ,line, c="g")
plt.xlabel("hours Studied")
plt.ylabel("Score in percentage")
plt.show()
```

STEP 6 - Testing out model by making predictions

```python
print(x_test) 
y_pred = lr.predict(x_test)
```

STEP 7 - Comparison of predicted result with the actual result

```python
CompareDf = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
CompareDf
```

STEP 8 - Testing the model with our own data

```python
hours =[[9.25]]
pred_value = lr.predict(hours)
print('Number of total hours : {}'.format(hours))
print('Predicted Score : {}'.format(pred_value[0]))
```

Thus by studing for 9 hours , a student can score 93.6917% .

STEP 9 - Calculating root mean squared error

```python
from sklearn import metrics
print('Mean Aboslute Error: ', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_pred))
print('Root mean squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R-2 :',metrics.r2_score(y_test, y_pred))
```

R-2 value of our model is 0.94549 which is a really great score
